```python
import pyvista as pv
from microgen import Tpms
from microgen.remesh import remesh_keeping_periodicity_for_fem
from microgen.shape.surface_functions import gyroid
```

```python
initial_gyroid = pv.UnstructuredGrid(
    Tpms(surface_function=gyroid, density=0.3, resolution=30).generateVtk(type_part="sheet")
)
```

```python
pv.start_xvfb()
initial_gyroid.plot(color="white", show_edges=True, screenshot="initial_gyroid.png")
```

```python
max_element_edge_length = 0.05
remeshed_gyroid = remesh_keeping_periodicity_for_fem(
    initial_gyroid, hmax=max_element_edge_length
)
```

```python
pv.start_xvfb()
remeshed_gyroid.plot(color="white", show_edges=True, screenshot="remeshed_gyroid.png")
```
